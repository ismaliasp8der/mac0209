### Testes com números reais

### Informações sobre o `float` em python

```python
epsilon = sys.float_info.epsilon
print("Diferença entre 1.0 e o menor valor representável em float que é maior que 1.0: epsilon =",
      epsilon)
print("Número de digitos binários na mantissa:", sys.float_info.mant_dig)
print("Número máximo de dígitos decimais que são confiavelmente representados:", sys.float_info.dig)

print("x = 3.1415926535897932 (pi com 16 casas decimais)")
x = 3.1415926535897932
print(f"x = {x:.16f}")
```

### Conceito de falsa precisão

```python
print("Simulando a medição de uma distância entre pontos.")

dist_AB = 205.10
dist_BC = 60.625
dist_CD = 750
dist_AD = dist_AB + dist_BC + dist_CD

print(f"Distância de A a B: {dist_AB:.2f}")
print(f"Distância de B a C: {dist_BC:.3f}")
print(f"Distância de C a D: {dist_CD:.0f}")
print(f"Distância de A a D calculada: {dist_AD:.3f}")

print("Aqui existe uma falsa precisão, pois estamos considerando uma precisão que não existe em todas as medições.")
print(f"Melhor representação: {dist_AD:.0f}")
```
