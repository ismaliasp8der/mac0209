# Setup

```python
# Setup aqui
import math
import matplotlib.pyplot as plt
import numpy as np
```

# Main

```python
# Exercício aqui
def x_next(x, r):
    return r * x * (1 - x)

def logistic_map(x0, r, N):
    ### 0 <= x0 <= 1
    ### 0 <= r  <= 4
    X = [x0]
    i = 1
    while(i<N):
        X.append(x_next(X[-1], r))
        i+=1
    return X
    
def plot_map(x0, r, N):
    X = logistic_map(x0, r, N)
    
    plt.figure(figsize=[9.6, 4.8])
    plt.plot(X, '+-c')
    plt.grid(ls='--', alpha=0.5)
    plt.title(r'Mapa Logístico: $x_0 = $ ' + str(x0) + r'; $r = $' + str(r))
    plt.ylabel(r'$x_n$')
    plt.xlabel(r'$n \in \mathbb{N}$')

#        x0    r    N
plot_map(0.9, 2.9, 100)
plot_map(0.4, 3.6, 200)

# Diagrama de Bifurcação
plt.figure(figsize=[9.6, 4.8])
plt.grid(ls='--', alpha=0.5)
plt.title(r'Diagrama de Bifurcação do Mapa Logístico')
plt.ylabel(r'$x_{200}$ para $x_0 \in (0,1)$')
plt.xlabel(r'$r$')

R = np.linspace(1,3.9,500)
for x in np.linspace(0.01,0.99,500):
    X = []
    for r in R:
        X.append(logistic_map(x, r, 200)[-1])
    plt.plot(R,np.array(X),', k')
```
