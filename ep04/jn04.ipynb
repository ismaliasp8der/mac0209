# Setup

```python
# Setup aqui
import math
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import numpy as np
```

# Main

```python
# Exercício aqui
def x_next(x, r):
    return r * x * (1 - x)

def logistic_map(x0, r, inicio, fim):
    ### 0 <= x0 <= 1
    ### 0 <= r  <= 4
    X = [x0]
    i = inicio
    while(i<fim):
        X.append(x_next(X[-1], r))
        i+=1

    plt.figure(figsize=[9.6, 4.8])
    plt.plot(X, '+-c')
    plt.grid(ls='--', alpha=0.5)
    plt.title(r'Mapa Logístico: $x_0 = $ ' + str(x0) + r'; $r = $' + str(r))
    plt.ylabel(r'$x_n$')
    plt.xlabel(r'$n \in \mathbb{N}$')

logistic_map(0.9, 2.9, 1, 100)
logistic_map(0.2, 3.6, 1, 200)
```
