# Main

```python
# Exercício aqui
def x_next(x, r):
    return r * x * (1 - x)

def logistic_map(x0, r, inicio, fim):
    ### 0 <= x0 <= 1
    ### 0 <= r  <= 4
    X = [x0]
    i = inicio
    while(i<fim):
        X.append(x_next(X[-1], r))
        i+=1

    plt.plot(X)
    plt.title(r'Mapa Logístico: $x_0 = $' ; r = 2.5$')
    plt.ylabel(r'$x_n$')
    plt.xlabel(r'$n \in \mathbb{N}$')
    plt.figure()
    
plt.title(r'Mapa Logístico: $x_0 = 0.5; r = 2.5$')
logistic_map(0.5, 2.5, 1, 200)
plt.title(r'Mapa Logístico: $x_0 = 0.9; r = 3.6$')
logistic_map(0.9, 3.6, 1, 200)
```
