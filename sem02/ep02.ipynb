# Main

Modelagem da equação (2.7):
$$\frac{d^2x(t)}{dt^2} = a.$$
### Método Analítico
Integramos a equação, obtendo:
$$\frac{dx(t)}{dt} = at + b$$
$$x(t) = \frac{a}{2}t^2 + bt + c.$$
### Método Numérico de Euler
Definimos $v(t) = \frac{dx(t)}{dt}$. Aproximando $x(t+\Delta t)$ pela série de Taylor até primeira ordem, temos:
$$x(t + \Delta t) = x(t) + \Delta t \frac{dx(t)}{dt} = x(t) + \Delta t v(t).$$
e para $v(t+\Delta t)$:
$$v(t + \Delta t) = v(t) + \Delta t \frac{dv(t)}{dt} = v(t) + \Delta t \frac{d^2x(t)}{dt^2} = v(t) + \Delta t a.$$ 

Ou seja, $x_{n+1} = x_n + \Delta t v_n$ e $v_{n+1} = v_n + \Delta t \cdot a$.

## Implementação

```python
# implementação aqui
def nextEuler(x,v,a,t,dt):
  x.append(x[-1]+v[-1]*dt)
  v.append(v[-1]+a*dt)
  t.append(t[-1]+dt)

def nextXa(x0,v0,a,t,l):
  l.append(x0+v0*t+a*t*t/2)

def nextVa(v0,a,t,l):
  l.append(v0 + a*t)

x0 = 0
v0 = -1
t0 = 0

a = 2
dt = 1e-2

npassos = 100

xAn = [x0]
vAn = [v0]

xEuler = [x0]
vEuler = [v0]
tEuler = [t0]

for i in range(npassos):
  nextEuler(xEuler,vEuler,a,tEuler,dt)
  nextXa(x0,v0,a,tEuler[-1],xAn)
  nextVa(v0,a,tEuler[-1],vAn)


plt.figure(0)
plt.plot(tEuler, xEuler, label='Euler', marker='o')
plt.plot(tEuler, xAn, label='Analítico', marker='')
plt.ylabel(r'posição $x(t)$')
plt.xlabel(r'tempo $t$')
plt.legend()
plt.title("Posição")
plt.show()

plt.figure(1)
plt.plot(tEuler, vEuler, label='Euler', marker='o')
plt.plot(tEuler, vAn, label='Analítico', marker='')
plt.ylabel(r'velocidade $v(t)$')
plt.xlabel(r'tempo $t$')
plt.legend()
plt.title("Velocidade")
plt.show()
```
