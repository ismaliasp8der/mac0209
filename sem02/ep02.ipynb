Modelagem da equação (2.7):
$$\frac{d^2x(t)}{dt^2} = a.$$
### Método Analítico
Integramos a equação, obtendo:
$$\frac{dx(t)}{dt} = at + b$$
$$x(t) = \frac{a}{2}t^2 + bt + c.$$
### Método Numérico de Euler
Definimos $v(t) = \frac{dx(t)}{dt}$. Aproximando $x(t+\Delta t)$ pela série de Taylor até primeira ordem, temos:
$$x(t + \Delta t) = x(t) + \Delta t \frac{dx(t)}{dt} = x(t) + \Delta t v(t).$$
e para $v(t+\Delta t)$:
$$v(t + \Delta t) = v(t) + \Delta t \frac{dv(t)}{dt} = v(t) + \Delta t \frac{d^2x(t)}{dt^2} = v(t) + \Delta t a.$$ 

Ou seja, $x_{n+1} = x_n + \Delta t v_n$ e $v_{n+1} = v_n + \Delta t \cdot a$.
